```python
import numpy as np
import pandas as pd

# Define the range of z values and the activation functions
z_values = np.arange(-1, 1, 0.3)
activation_functions = {
    "None": lambda z: z,
    "Sigmoid": lambda z: 1 / (1 + np.exp(-z)),
    "Tanh": lambda z: (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z)),
    "ReLU": lambda z: np.maximum(0, z)
}

# Create a DataFrame to hold the results
result_df = pd.DataFrame(index=z_values, columns=activation_functions.keys())

# Compute the output for each activation function and z value
for z in z_values:
    for function_name, function in activation_functions.items():
        result_df.loc[z, function_name] = function(z)

result_df
```
