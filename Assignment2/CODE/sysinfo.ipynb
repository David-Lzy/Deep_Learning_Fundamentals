```python
import subprocess


# 查询CPU信息
def get_cpu_info():
    try:
        result = subprocess.run(["lscpu"], capture_output=True, text=True, check=True)
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error retrieving CPU info: {e}"


# 查询内存信息
def get_memory_info():
    try:
        result = subprocess.run(
            ["free", "-h"], capture_output=True, text=True, check=True
        )
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error retrieving memory info: {e}"


# 查询显卡信息
def get_gpu_info():
    try:
        result = subprocess.run(
            ["nvidia-smi"],
            capture_output=True,
            text=True,
            check=True,
        )
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error retrieving GPU info: {e}"


# 查询操作系统信息
def get_os_info():
    try:
        result = subprocess.run(
            ["lsb_release", "-a"], capture_output=True, text=True, check=True
        )
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error retrieving OS info: {e}"


if __name__ == "__main__":
    print("CPU Info:")
    print(get_cpu_info())

    print("\nMemory Info:")
    print(get_memory_info())

    print("\nGPU Info:")
    print(get_gpu_info())

    print("\nOS Info:")
    print(get_os_info())
```
