```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load all the data
directory_path = "./L"
files = [1, 2, 4, 8, 16]
dataframes = {}

for file in files:
    dataframes[file] = pd.read_csv(f"{directory_path}/{file}.csv")

# Plotting
plt.figure(figsize=(12, 12))
plt.xlim(0, 100)
# plt.ylim(0.5, df["Training Accuracy"].max())
sns.set_style("whitegrid")

# plt.yscale("log")
# sns.set(font_scale=1.4)

for key, df in dataframes.items():
    sns.lineplot(x=df["Epoch"], y=df["Training Accuracy"], label=f"L={key}")

plt.title("Training Accuracy over Epochs for Different Attention Heads (L)")
plt.xlabel("Epoch")
plt.ylabel("Training Accuracy")
plt.legend(title="Attention Heads (L)")
plt.tight_layout()
plt.show()
```

```python
# Plotting the relationship between epoch and testing accuracy with all attention head configurations
plt.figure(figsize=(12, 12))
plt.xlim(0, 100)
sns.set_style("whitegrid")

# Plotting all the attention head configurations
for key, df in dataframes.items():
    sns.lineplot(x=df["Epoch"], y=df["Testing Accuracy"], label=f"L={key}")

plt.title("Testing Accuracy over Epochs for Different Attention Heads (L)")
plt.xlabel("Epoch")
plt.ylabel("Testing Accuracy (%)")
plt.legend(title="Attention Heads (L)")
plt.tight_layout()
plt.show()
```

```python
# Extracting final epoch results for each attention head configuration
final_results = []

for key, df in dataframes.items():
    final_results.append(df.iloc[-1])

# Creating a DataFrame for the final results
final_df = pd.DataFrame(final_results)
final_df = final_df.reset_index(drop=True)
final_df["Attention Heads (L)"] = dataframes.keys()
final_df = final_df.set_index("Attention Heads (L)")
final_df
```

```python
# Plotting Evaluation Time vs Epoch for different attention heads
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

for key, df in dataframes.items():
    sns.lineplot(x=df["Epoch"], y=df["Evaluation Time"], label=f"L={key}")

plt.xscale("log")
plt.yscale("log")
plt.title("Evaluation Time over Epochs for Different Attention Heads (L)")
plt.xlabel("Epoch")
plt.ylabel("Evaluation Time (log scale)")
plt.legend(title="Attention Heads (L)")
plt.tight_layout()
plt.show()
```
