```python
import os
import pandas as pd


def process_csv_files(directory):
    # 创建一个空的DataFrame来存储结果
    result_df = pd.DataFrame()

    # 遍历指定目录下的所有CSV文件
    for file in os.listdir(directory):
        if file.endswith(".csv"):
            # 构造完整的文件路径
            file_path = os.path.join(directory, file)

            # 读取CSV文件的最后一行
            df = pd.read_csv(file_path)
            last_row = df.iloc[-1]

            # 将文件名和最后一行数据添加到结果DataFrame中
            result_df = result_df.append(last_row, ignore_index=True)
            result_df.at[result_df.index[-1], "File Name"] = file

    return result_df


# 指定要处理的目录
directory = "/Project/David_Li/Works/Courses/4339_COMP_SCI_7318_Deep_Learning_Fundamentals/Assignment3/DATA/Result/1m"  # 替换为你的目录路径

# 处理CSV文件并获取结果
result_df = process_csv_files(directory)

# 打印结果
print(result_df)
```
